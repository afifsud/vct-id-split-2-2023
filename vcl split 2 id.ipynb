```python
#mendefinisikan data ke chart
x = df["Impact"]
y = df["FBSR%"]
z = df["Names"]

#membuat chart nya
fig = plt.figure(figsize=(16,8))
ax =fig.add_subplot(111)

#membuat label tiap data point
for i, label in enumerate(z):
    plt.text(x[i], y[i], label, fontsize=10, ha='right')

plt.scatter(x, y)
plt.xlabel('Impact (Right More Better)')
plt.ylabel('FBSR% (Upper Better)')
plt.grid(True)


plt.show()
```
